```python
# Import required modules
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
# Read csv file
file_path = r"C:\Users\hankc\OneDrive\桌面\py4e\ebay_Laptop_Product_.csv"
data = pd.read_csv(file_path)
data.head(5)
```

```python
# Data observation
# Observe data types, number of data entries, and number of columns
# Check whether there are duplicate rows or missing values
info = data.info()
shape = data.shape
missing_values = data.isnull().sum()
has_duplicates = data.duplicated().any()

print("Data Info:")
print(info)

print("\nDataFrame Shape:")
print(shape)

print("\nMissing Values:")
print(missing_values)

print("\nAre there duplicate rows?")
print(has_duplicates)
```

```python
# Remove duplicate rows
unique_data = data.drop_duplicates()
has_duplicates = unique_data.duplicated().any()

print("\nAre there duplicate rows?")
print(has_duplicates)
```

```python
# Data type transformation
# Convert specific columns to string data types
unique_data['Product_Price'] = unique_data['Product_Price'].astype(str)
unique_data['Place_of_shipment'] = unique_data['Place_of_shipment'].astype(str)
unique_data['Review'] = unique_data['Review'].astype(str)
unique_data['Shipping_Cost'] = unique_data['Shipping_Cost'].astype(str)
```

```python
# Data cleaning
# Remove specific symbols and words
unique_data['Product_Price'] = unique_data['Product_Price'].str.replace('NT$', '').str.replace(',', '')
unique_data['Place_of_shipment'] = unique_data['Place_of_shipment'].str.replace('from', '')
unique_data['Review'] = unique_data['Review'].str.replace(',', '')
unique_data['Shipping_Cost'] = unique_data['Shipping_Cost'].str.replace('\+NT\$|\s*shipping', '', regex=True).str.replace(',', '')

unique_data.head(5)
```

```python
# Data cleaning
# Define a function to process ['Product_Price'] containing price ranges, converting them into average prices and then into integers
# Maintain the original format for other types of prices and convert them into integers
def calculate_average_price(price_str):
    try:
        # Check if 'to' is in the ['Product_Price']
        if ' to ' in price_str:
            # Split the price range string, remove spaces and commas, and convert to integers
            prices = [int(price.replace(',', '').strip()) for price in price_str.split(' to ')]
            if len(prices) == 2:
                # Return the average price of the range
                return sum(prices) // 2
            else:
                return None
        else:
            # Price string without 'to', convert to integer directly
            price = int(price_str.replace(',', '').strip())
            return price
    except ValueError as ve:
        print(f"Error: Invalid price format in '{price_str}': {ve}")
        return None
    except Exception as e:
        print(f"Error processing price {price_str}: {e}")
        return None
        
unique_data['Product_Price'] = unique_data['Product_Price'].apply(calculate_average_price)
unique_data.head(5)
```

```python
# Data cleaning
# Define a function to process ['Review'], extracting strings in brackets , converting them into integers
def extract_string_in_brackets(text):
    # Use regular expressions to match strings within brackets
    match = re.search(r'\(([^)]+)\)', text)  
    if match:
        # Return the string in brackets
        return match.group(1)  
    else:
        return None 

unique_data['Review'] = unique_data['Review'].apply(extract_string_in_brackets)
unique_data['Review'] = unique_data['Review'].astype('int64')
unique_data.head(5)
```

```python
# Data observation after data cleaning
info = unique_data.info()
shape = data.shape
missing_values = data.isnull().sum()


print("Data Info:")
print(info)

print("\nDataFrame Shape:")
print(shape)

print("\nMissing Values:")
print(missing_values)
```

```python
# Convert data to numeric type and replace missing values ​​with 0
unique_data['Shipping_Cost'] = pd.to_numeric(unique_data['Shipping_Cost'], errors='coerce').fillna(0).astype('int64')
info = unique_data.info()

print("Data Info:")
print(info)
```

```python
output_file_path = r"C:\Users\hankc\OneDrive\桌面\py4e\cleaned_ebay_Laptop_Product.csv"
unique_data.to_csv(output_file_path, index=False)
```
